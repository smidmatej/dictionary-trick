```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def logistic(x, r):
    return r*x* (1-x)
```

```python
x = np.linspace(0.001, 0.999 ,100)
r_all = np.linspace(1, 4 , 10000)
y = np.empty((len(x), len(r_all)))

iterations = 100


for ir, r in zip(range(len(r_all)), r_all):
    y[:,ir] = logistic(x, r)
    for t in  range(iterations):
        y[:, ir] = logistic(y[:, ir], r)

fig = plt.Figure(figsize=(2,2))
plt.plot(x,y[:, 0])
```

```python
fig = plt.Figure(figsize=(10,10), dpi=100)
for ix in range(len(x)):
    plt.plot(r_all, y[ix, :], ',k', alpha=.25)
plt.savefig("logistic_map.png")
plt.show()
```

```python
x = 0.8
r_all = np.linspace(1, 4 , 1000)
iterations = 1000
y = np.empty((len(r_all), iterations))




for ir, r in zip(range(len(r_all)), r_all):
    y[ir, 0] = logistic(x, r)
    for t in  range(1, iterations):
        y[ir, t] = logistic(y[ir, t-1], r)

fig = plt.Figure(figsize=(2,2))
plt.plot(r_all, y, ',k', alpha=.25)

```

```python
r_chaos = 3.6


n = 10000
y = np.empty((n,))
y[0] = 0.5

for i in range(1, n):
    y[i] = logistic(y[i-1], r_chaos)

fig = plt.Figure(figsize=(2,2))
plt.plot(range(n), y)
```

The system is not ergodic? Does ergodicity mean that the system visits all points in the state space with the same probability? Probably not. Its more likely that ergodicity means that the system visits all points in the state space with non zero probability, ie. no place is unvisited after infinite amount of iterations
